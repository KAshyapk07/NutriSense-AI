```python

df1['clean_name'] = df1['Dish Name'].str.lower().str.strip()
df2['clean_name'] = df2['TranslatedRecipeName'].str.lower().str.strip()

df1['clean_name'] = df1['clean_name'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
df2['clean_name'] = df2['clean_name'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
set1 = set(df1['clean_name'])
set2 = set(df2['clean_name'])

common = set1.intersection(set2)
only_in_df1 = set1 - set2
only_in_df2 = set2 - set1

print("Common dishes:", len(common))
print("Nutrition dataset only:", len(only_in_df1))
print("Recipe dataset only:", len(only_in_df2))
```
